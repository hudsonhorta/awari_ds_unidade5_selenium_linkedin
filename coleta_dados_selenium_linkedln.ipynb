```python
# importação de pacotes e bibliotecas necessárias

from selenium import webdriver  # uso da biblioteca selenium com webdriver para fazer o Web Scraping
from time import sleep # uso da função sleep da time para aguardar tempo de carregamento da página
import os # uso da os para tratar o arquivo texto

# funcao para verificar se o arquivo texto existe e apaga o mesmo
def apaga_arquivo():
    path = 'C:\\Users\\hudso'
    dir = os.listdir(path)
    for file in dir:
        if file == arquivo_texto:
            os.remove(file)

# arquivo texto para armazenar as vagas
arquivo_texto = 'descricao_vagas.txt'

# criacao da instancia do webdriver
driver = webdriver.Chrome()

# url do Linkdin para pegar os dados de vagas
url = 'https://www.linkedin.com/jobs/cientista-de-dados-jobs?originalSubdomain=br&countryRedirected=1&position=1&pageNum=0'

# realizar a abertura do navegador e leitura da pagina
driver.get(url)
# aguarda 3 segundos para total carregamento da pagina
sleep(3)

# procura o modal que é apresentado quando a página é aberta a primeira vez e clica para fechar o modal
modal = driver.find_element_by_class_name('cta-modal__dismiss-btn')
modal.click()

# chama função para apagar o arquivo texto se existir
apaga_arquivo()

# faz um loop para ler as primeiras 5 vagas disponiveis
for x in range(5):
    # caminha da vaga de acordo com cada x do loop
    caminho = '//*[@id="main-content"]/div/section[2]/ul/li['+str(x+1)+']/a'
    
    # procura a vaga e da um click e agurda 1 segundo para carregamento da página
    cargo = driver.find_element_by_xpath(caminho)
    cargo.click()
    sleep(1)
    
    # faz leitura dos dados necessários
    titulo = driver.find_element_by_xpath('//*[@id="main-content"]/section/div[2]/section[1]/div[1]/div[1]/a/h2')
    empresa = driver.find_element_by_class_name('topcard__flavor')
    print(f'Empresa encontrada: {empresa.text}')
    descricao = driver.find_element_by_class_name('description')

    # salva as informações lidas no arquivo texto
    with open(arquivo_texto, 'a', encoding='utf-8') as arquivo:
        arquivo.write('Cargo: ' + titulo.text + '\n' + 'Empresa: ' + empresa.text + '\n\n'  + descricao.text)
        arquivo.write('\n----------------------------\n')

# informa final do processo
print(f'Arquivo [{arquivo_texto}] criado com as ultimas 5 vagas no LinkEdln sobre Data Science!')
```
